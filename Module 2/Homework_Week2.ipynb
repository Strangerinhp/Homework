```python
import cv2
import numpy as np

def compute_difference(bg_img, input_img):
    difference = cv2.absdiff(bg_img, input_img)
    return cv2.cvtColor(difference, cv2.COLOR_BGR2GRAY)

def compute_binary_mask(difference_single_channel):
    _, binary_mask = cv2.threshold(difference_single_channel, 30, 255, cv2.THRESH_BINARY)
    return binary_mask

def replace_background(bg1_image, bg2_image, ob_image):
    difference_single_channel = compute_difference(bg1_image, ob_image)
    binary_mask = compute_binary_mask(difference_single_channel)
    output = np.where(binary_mask == 255, ob_image, bg2_image)
    return output
```

```python
import numpy as np

def inverse_matrix(matrix):
    return np.linalg.inv(matrix)
```

```python
def compute_eigenvalues_eigenvectors(matrix):
    eigenvalues, eigenvectors = np.linalg.eig(matrix)
    # Normalize eigenvectors
    eigenvectors = eigenvectors / np.linalg.norm(eigenvectors, axis=0)
    return eigenvalues, eigenvectors
```

```python
def compute_cosine(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def load_and_resize_image(path, size=(678, 381)):
    img = cv2.imread(path)
    return cv2.resize(img, size)

def compute_difference(bg_img, input_img):
    difference = cv2.absdiff(bg_img, input_img)
    return cv2.cvtColor(difference, cv2.COLOR_BGR2GRAY)

def compute_binary_mask(difference_single_channel, threshold=30):
    _, binary_mask = cv2.threshold(difference_single_channel, threshold, 255, cv2.THRESH_BINARY)
    return binary_mask

def replace_background(bg1_image, bg2_image, ob_image):
    difference_single_channel = compute_difference(bg1_image, ob_image)
    binary_mask = compute_binary_mask(difference_single_channel)
    output = np.where(binary_mask[:,:,np.newaxis] == 255, ob_image, bg2_image)
    return output.astype(np.uint8)

# Load and resize images
bg1_image = load_and_resize_image('GreenBackground.png')
ob_image = load_and_resize_image('Object.png')
bg2_image = load_and_resize_image('NewBackground.jpg')

# Compute difference
difference_single_channel = compute_difference(bg1_image, ob_image)

# Compute binary mask
binary_mask = compute_binary_mask(difference_single_channel)

# Replace background
output = replace_background(bg1_image, bg2_image, ob_image)

# Display results
cv2_imshow(bg1_image)
cv2_imshow(ob_image)
cv2_imshow(bg2_image)
cv2_imshow(difference_single_channel)
cv2_imshow(binary_mask)
cv2_imshow(output)

# Save output image
cv2.imwrite('output.png', output)
```
