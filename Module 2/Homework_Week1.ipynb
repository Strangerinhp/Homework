```python
! gdown 1iA0WmVfW88HyJvTBSQDI5vesf-pgKabq
import pandas as pd
df = pd.read_csv('/content/advertising.csv')
data = df.to_numpy()
max_sale = np.max(data[:,3])
max_sale_arg = np.argmax(data[:,3])
print(max_sale)
print(max_sale_arg)
```

```python
mean_tv = np.mean(data[:,0])
print(mean_tv)
```

```python
high_sale = data[data[:,3] >= 20]
print(high_sale)
print(high_sale.shape)
```

```python
mean_of_high_radio = np.mean(data[data[:,-1]>15],)
print(mean_of_high_radio)
```

```python
mask = data[:, -1] >= 15
radio_mean = np.mean(data[mask, 1])
print(radio_mean)
```

```python
mean_news = np.mean(data[:,2])
sum_good_news = np.sum(data[data[:,2] >= mean_news,-1])
print(sum_good_news)
```

```python
sales_mean = np.mean(data[:, -1])
scores = np.where(data[:, -1] > sales_mean, 'Good',
                  np.where(data[:, -1] < sales_mean, 'Bad', 'Average'))
print(scores[7:10])
```

```python
sales_mean = np.mean(data[:, -1])
closest_to_mean = data[np.argmin(np.abs(data[:, -1] - sales_mean)), -1]
scores = np.where(data[:, -1] > closest_to_mean, 'Good',
                  np.where(data[:, -1] < closest_to_mean, 'Bad', 'Average'))
print(scores[7:10])
```
