```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.preprocessing import OrdinalEncoder , StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error , mean_squared_error
```

```python
housing = pd.read_csv('/content/Housing.csv')
housing.head()
```

```python
categorical_cols = housing.select_dtypes(include=['object']).columns.to_list()
print(categorical_cols)
```

```python
ordinal_encoder = OrdinalEncoder()
encoded_categorical_cols = ordinal_encoder.fit_transform(housing[categorical_cols])
encoded_categorical_df = pd.DataFrame(encoded_categorical_cols, columns=categorical_cols)
numerical_df = housing.drop(columns=categorical_cols)
encoded_df = pd.concat([numerical_df, encoded_categorical_df], axis = 1)
encoded_df.head()
```

```python
normalizer = StandardScaler()
dataset_arr = normalizer.fit_transform(encoded_df)
```

```python
X, y = dataset_arr[:, 1:], dataset_arr[:,0]
```

```python
test_size = 0.3
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=1, shuffle = True)
```

```python
regressor = GradientBoostingRegressor(random_state=1)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
mae = mean_absolute_error(y_test , y_pred )
mse = mean_squared_error(y_test, y_pred )

print('Evaluation results on validation set :')
print(f'Mean Absolute Error : {mae}')
print(f'Mean Squared Error : {mse}')
```

```python
RF = RandomForestRegressor(random_state=1)
RF.fit(X_train, y_train)
```

```python
y_pred_RF = RF.predict(X_test)
```

```python
mae = mean_absolute_error(y_test , y_pred_RF )
mse = mean_squared_error(y_test, y_pred_RF )

print('Evaluation results on validation set :')
print(f'Mean Absolute Error : {mae}')
print(f'Mean Squared Error : {mse}')
```
